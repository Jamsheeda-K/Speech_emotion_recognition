```python
import os
import librosa
import math
import json
import numpy as np
from sklearn.model_selection import train_test_split
import os
os.environ["CUDA_VISIBLE_DEVICES"]="-1"
import tensorflow as tf
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
sampling_rate = 22050
```

```python
trained_model = tf.keras.models.load_model('mfcc_only_model_calm_happy_angry_new.h5')
```

```python
def predict(model,X):
    #X = X[np.newaxis,...] #size of X (1,130,13,1) [Conversion to be done if entering one value]
    prediction = model.predict(X) #prediction contains 3 values - probability for each genre [[0.1, 0.2, ..]]
    #print(prediction)
    #extract index with max value
    predicted_index = np.argmax(prediction,axis=1)
    return predicted_index
```

```python
#predicting the emotion from external audio
# processing the input audio
my_test_audio,sr =librosa.load('angry_set.wav',sr=sampling_rate)
audio_duration = round(librosa.get_duration(my_test_audio),0)
std_duration = 3
if audio_duration < std_duration:
    print("The audio should be atleast 3 seconds long")
```

```python
#play the audio
Audio(data=my_test_audio,rate=sampling_rate)
```

```python
#Extracting input features
mfcc_test = librosa.feature.mfcc(my_test_audio[:sampling_rate*(std_duration)],sr=sampling_rate,n_mfcc=13,n_fft=2048,hop_length=512)
mfcc_test = mfcc_test.T
mfcc_test = mfcc_test[np.newaxis,...,np.newaxis]
```

```python
#Making predictions
y_pred = predict(trained_model,mfcc_test)
Emotions = {0:'calm', 1:'happy',2:'angry'}
Emotions[y_pred[0]]
```
